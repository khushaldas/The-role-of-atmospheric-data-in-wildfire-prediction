```python
CAL_2021_Data_path = '/content/drive/MyDrive/Forest_Fire_Data/updated/CAL_2021_complete.csv'

# Read the csv using pandas
CAL_2021_Data = pd.read_csv(CAL_2021_Data_path)
```

```python
CAL_2021_Data.head()
```

```python
columns_to_remove=['Ent_Counter','DDec','dVAL']

CAL_2021_Data = CAL_2021_Data.drop(columns=columns_to_remove)
```

```python
CAL_2021_Data.head()
```

```python
# Combine date and time columns into a single datetime column
CAL_2021_Data['datetime_column'] = pd.to_datetime(CAL_2021_Data['date'] + ' ' + CAL_2021_Data['time'])
```

```python
CAL_2021_Data.head()
```

```python
# Convert each combined datetime value to a decimal point representing days with minutes and seconds
CAL_2021_Data['decimal_time'] = ((CAL_2021_Data['datetime_column'] - pd.to_datetime(CAL_2021_Data['datetime_column'].dt.year, format='%Y')) / pd.Timedelta(days=1)).round(6)
```

```python
CAL_2021_Data.head()
```

```python
CAL_2021_Data = CAL_2021_Data.drop(columns=['time','date'])
```
